```python
import os
import sys
sys.path.append('/'.join(os.getcwd().split('/')[:-1]+['myPackages']))
import pandas as pd

import coffee_beans as cb
```

```python
%time
beans = {}

# Voila
beans['voila'] = cb.getVoila()

# Maruyama
beans['maruyama'] = cb.getMaruyama()
```

```python
beans_df = pd.concat([beans['maruyama'],
                      beans['voila']
                     ])
```

```python
flavors = ['マスカット', 'リンゴ',  'ライム', 'オレンジ', '巨峰', 'アプリコット', 'ライチ', 'プラム', '黄桃',
           'グレープフルーツ', 'レーズン', 'ジャスミン', 'ラベンダー', '白桃', 'いちご', 'ブルーベリー',
           '蜂蜜', 'アーモンド', '黒糖',
           'レモンティー', 'レモネード', 
           'ダークチョコ', 'ミルクチョコ', 'スイートポテト', 'ビターキャラメル',
          ]
#            'フローラル', '甘さ', 'フルーティー', 'すっきり', '爽やか',
#            '滑らかな口当たり', '丸みのある口当たり'


flavor_cnt =  [[flavor, ' '.join([value for value in beans_df['detial'].values]).count(flavor)] for flavor in flavors]
```

```python
top_flavors = sorted(flavor_cnt, key=lambda x: -x[1])[:5]
top_flavors = [i[0] for i in top_flavors]
print(top_flavors)
beans_df['flavors'] = beans_df['detial'].apply(lambda x:[y for y in [tf if tf in x else None for tf in top_flavors] if y])
```

```python
# format 
beans_df['shop'] = beans_df['shop'].replace({'maruyama': '丸山珈琲',
                                             'voila': 'ヴォアラコーヒー'})

# rm row
## 同一商品
beans_df = beans_df.sort_values('gram')
beans_df = beans_df.drop_duplicates(subset='name')

# add col
beans_df['15g'] = beans_df['1g']*15
```

```python
beans_df.loc[:,['shop', 'name', 'value', 'gram', 'country','roast', 'detial', 'flavors', 'url']]
```

```python
beans_json = beans_df.loc[:,
                          ['shop', 'name', 'value', 'gram', 'country','roast', 'detial', 'flavors', 'url']
                         ].to_json(orient='records', force_ascii=False)
```

```python
with open('beans.json', 'w')as f:
    f.write(beans_json[1:-1])
```
